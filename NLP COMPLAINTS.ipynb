```python
a=df[df['category']==0].sample(n=10000,random_state=0)
b=df[df['category']==1].sample(n=10000,random_state=0)
c=df[df['category']==2].sample(n=10000,random_state=0)
d=df[df['category']==3].sample(n=10000,random_state=0)
e=df[df['category']==4].sample(n=10000,random_state=0)
f=df[df['category']==5].sample(n=10000,random_state=0)
g=df[df['category']==6].sample(n=10000,random_state=0)
h=df[df['category']==7].sample(n=10000,random_state=0)
```

```python
dfsample.index=np.arange(0,80000)
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem.lancaster import LancasterStemmer
```

```python
unw=set(stopwords.words('english')+list(punctuation))
```

```python
stemmer = LancasterStemmer()
```

```python
corpus=dfsample['Consumer complaint narrative']
```

```python
corpus
```

```python
final_corpus= []
for i in range(len(corpus)):
    text = word_tokenize(corpus[i].lower())
    text = [stemmer.stem(x) for x in text if x not in unw]
    final = " ".join(text)
    final_corpus.append(final)
```

```python
final_corpus
```

```python
new=pd.DataFrame(final_corpus)
```

```python
new['target']=dfsample['category']
```

```python
new
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
```

```python
vector= tfidf.fit_transform(new[0])
```

```python
X = vector.toarray()
y = new['target']
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
```

```python
mnb.fit(X_train,y_train)
```

```python
y_pred=mnb.predict(X_test)
```

```python
y_pred_train=mnb.predict(X_train)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
confusion_matrix(y_train,y_pred_train)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(classification_report(y_train,y_pred_train))
```
